#Average Scores Comparison of Male And Female

```python
import pandas as pd
df = pd.read_excel("/content/Day15_Statistics_Practice.xlsx")

def result():
  # Calculate the mean math score for males
  male_math_score_mean = df[df['gender'] == 'male']['math_score'].mode()

  # Calculate the mean math score for females
  female_math_score_mean = df[df['gender'] == 'female']['math_score'].mode()

  if male_math_score_mean == female_math_score_mean:
    print("Both scores are equal")
  elif male_math_score_mean > female_math_score_mean:
    print("Male scores are higher")
  else:
    print("Female scores are higher")

result()
```

```python
def male():
  male_score = df[df['gender']=="male"] ['math_score'].mean()
  female_score = df[df['gender']=="female"]['math_score'].mean()
  if male_score==female_score:
    print("Both scores are equal")
  elif male_score> female_score:
    print("Male scores are higher")
  else:
    print("Female scores are higher")

male()
```

```python
df["gender"] = df['gender'].str.replace("male", "M")
df["gender"] =df["gender"].str.replace("female", "F")
df
```
