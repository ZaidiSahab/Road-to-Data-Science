```python
import pandas as pd
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isna().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
labels = ['Young Adult','Adult','Middle Age','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels = labels)
```

```python
df[['age_group','age']].head(10)
```

```python
#create coulmn Purchase_frequency_Days
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly' :30 ,
    'Quaterly' :90,
    'Bi-Weekly' :14,
    'Annually' :365,
    'Every 3 months' :90}


```

```python
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].tail(10)
```

```python
df[['promo_code_used' , 'discount_applied']].head(10)
```

```python
(df['promo_code_used']== df['discount_applied']).all()
```

```python
df = df.drop('promo_code_used',axis =1)
```
